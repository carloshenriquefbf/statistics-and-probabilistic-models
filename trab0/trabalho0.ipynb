```python
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt
```

# Importando os dados

```python
train_data = pd.read_csv('./train.csv')
test_data = pd.read_csv('./test.csv')

merged_data = pd.concat([train_data, test_data], ignore_index=True)
```

```python
print(merged_data.describe())
```

# Limpando o dataset e utilizando apenas 3 colunas

```python
data = merged_data[['Age', 'Height', 'Weight']]
```

# Cálculo da média

```python
avg_values = data.mean()
print("Average Values:")
print(avg_values)
```

# Cálculo da mediana

```python
median_values = data.median()
print("Median Values:")
print(median_values)
```

# Cálculo do desvio padrão

```python
std_values = data.std()
print("Standard Deviation Values:")
print(std_values)
```

# Cálculo do coeficiente de correlação

```python
correlation_matrix = data.corr()
print("Correlation Matrix:")
print(correlation_matrix)
sns.heatmap(correlation_matrix, annot=True, cmap='coolwarm')
plt.title('Correlation Heatmap')
plt.show()
```

# Visualização dos dados

```python
plt.figure(figsize=(12, 6))
plt.subplot(1, 3, 1)
sns.boxplot(y=data['Age'])
plt.title('Box Plot of Age')
plt.subplot(1, 3, 2)
sns.boxplot(y=data['Height'])
plt.title('Box Plot of Height')
plt.subplot(1, 3, 3)
sns.boxplot(y=data['Weight'])
plt.title('Box Plot of Weight')
```

```python
plt.figure(figsize=(12, 6))
plt.subplot(1, 3, 1)
sns.histplot(data['Age'], kde=True)
plt.title('Histogram of Age')
plt.subplot(1, 3, 2)
sns.histplot(data['Height'], kde=True)
plt.title('Histogram of Height')
plt.subplot(1, 3, 3)
sns.histplot(data['Weight'], kde=True)
plt.title('Histogram of Weight')
```
